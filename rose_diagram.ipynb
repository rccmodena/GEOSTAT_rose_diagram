# Rose Diagram

Create a Rose Diagram with the principal directions of the Walker Lake sample dataset (ISAAKS; SRIVASTAVA, 1989, p. 153).

```python
import math
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Ellipse

%matplotlib inline

tamanhoEixos = 45
linhas = [[33.89,90,'0.4'], [19.43,67.5,'0.4'], [14.74,45,'0.4'], [13.04,22.5,'b'], [17.72,0,'0.4'], 
[27.09,157.5,'0.4'], [32.19,135,'0.4'], [39,112.5,'r']]

fig = plt.figure(figsize=(10, 10), dpi=80, facecolor='w')
ax = plt.axes((0, 0, 1, 1), xticks=[], yticks=[], frameon=False)

ax.annotate('', (-tamanhoEixos, 0), (tamanhoEixos, 0),
            ha='center', va='center',
            arrowprops=dict(arrowstyle='<->', color='k', lw=1),zorder = 1)
ax.annotate('', (0, -tamanhoEixos), (0, tamanhoEixos),
            ha='center', va='center',
            arrowprops=dict(arrowstyle='<->', color='k', lw=1),zorder = 1)

ax.annotate('N', xy=(0, 0), xytext=(-1, 45))
ax.annotate('L', xy=(0, 0), xytext=(45, -1))
ax.annotate('O', xy=(0, 0), xytext=(-48, -1))
ax.annotate('S', xy=(0, 0), xytext=(-1, -48))

ax.annotate(u'337.5°', xy=(0, 0), xytext=(-22, 40))
ax.annotate(u'157.5°', xy=(0, 0), xytext=(14, -42))
ax.annotate(u'67.5°', xy=(0, 0), xytext=(16, 6))
ax.annotate(u'247.5°', xy=(0, 0), xytext=(-24, -8))

for tam, ang, cor in linhas:
    ax.arrow(0, 0, np.cos(ang * 2 * math.pi / 360.0)*tam,
             np.sin(ang * 2 * math.pi / 360.0)*tam, head_width=1, head_length=1, fc=cor,
             ec=cor)
    ax.arrow(0, 0, np.cos((ang+180) * 2 * math.pi / 360.0)*tam,
             np.sin((ang+180) * 2 * math.pi / 360.0)*tam, head_width=1, head_length=1,
             fc=cor , ec=cor)

rotation = 112.5
ax.add_patch(Ellipse((0, 0), 2* 40, 2 * 14.04, rotation,
ec='k', fc='gray', alpha=0.2, zorder=1))

ax.set_xlim(-60, 60)
ax.set_ylim(-60, 60)

plt.show()
```
